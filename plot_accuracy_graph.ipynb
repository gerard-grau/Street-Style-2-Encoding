```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
submission_scores = {
    'idx':          [1,         2,          3,          5,          12,         16,         20],
    'time':         [-19,       -18,        -17,        -15,        -10,        -5,         -1],
    'kaggle_score': [0.72749,   0.77109,    0.77162,    0.78025,    0.81084,    0.81259,    0.81392],
    'labels':       ['KNN', 'XGBoost', '', '', 'XGB + metadata', '', 'XGB + tune hyperparameters' ]
}
```

```python
df = pd.DataFrame(submission_scores)
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Create the plot
plt.figure(figsize=(10, 6))
sns.lineplot(x=df['time'], y=df['kaggle_score'], marker='o', linestyle='-', color='b', label="Our submission scores' graph")
for i in range(len(df)):
    plt.text(df['time'][i], df['kaggle_score'][i], df['labels'][i], horizontalalignment='right', verticalalignment='bottom', size='small', color='#505050')

plt.axhline(0.78324, c='r', linestyle=':', label='2nd best competitor')
plt.title('Kaggle Submission Scores Over Time')
plt.xlabel('')
plt.ylabel('Kaggle Score')
plt.xlim(-24, 1)
custom_ticks = [-22, -13, -0.5]
custom_labels = ['Start', 'Midnight', 'End']
plt.xticks(custom_ticks, custom_labels)
plt.legend()
plt.tight_layout()
plt.savefig('kaggle_submission_scores.png')
```
